```python
'''
companyDat=pd.read_csv("Company.csv")
DF=pd.read_csv("CompanyValues.csv")
print(companyDat)
DF.head()

companies=DF.groupby(DF.ticker_symbol)

tesla= [comp for comp in companies if comp[0]=="TSLA"]
tesla[0][1].plot(x="day_date")
'''
#lets use yfinance
ticker = yfinance.Ticker("TSLA")
tesla = ticker.history(interval="1d",start="2015-03-15",end="2020-09-10")
#has colnames ["Date", "Open", "High", "Low", "Close", "Volume", "Dividends", "Stock Splits"]
```

## we have tesla data, lets visualize it

```python
plt.plot(tesla["Close"])
plt.xlabel("Time")
plt.ylabel("Closing price")
plt.show()
```

## the closing price is uh, pretty upwards
## lets do some stat stuff

```python
#start by doing everything in terms of closing prices
closing=tesla["Close"]
# daily returns = percent change in closing price between two days
    #theres a percent change method
returns=closing.pct_change(1).dropna() 
#the 1 is for the gap between entries that signify a unit
#drop na bc no null stuff 
#plt.plot(returns)
```

## returns is a mess, lets look at distribution of returns

```python
plt.hist(returns,bins="rice",label="Daily close price")
plt.legend()
plt.show()
```

## looks pretty centered around 0 but lets boxplot it just to see

```python
plt.boxplot(returns,labels=["Daily close price"])
plt.show()
```

```python
#thats got fat tails and a hella skinny inner quartile
#lets look at some vars
print(f'mean: {np.mean(returns)}\nstd: {np.std(returns)}\nskew: {skew(returns)}\nkutosis: {kurtosis(returns)}')
```

mean>0 makes sense bc the closing graph went up
std>mean, makes sense considering the massive outliers and mess of a returns graph
skew isnt 0, so returns are nonsymettrical. skew>0 so right tail is significant

## now for the important question on returns
### is returns normally ditributed 

```python
print(skewtest(returns)) #low pval means skew shouldnt be neglected
print(kurtosistest(returns))
```

    If skewness is less than -1 or greater than 1, the distribution is highly skewed.
    If skewness is between -1 and -0.5 or between 0.5 and 1, the distribution is moderately skewed.
    If skewness is between -0.5 and 0.5, the distribution is approximately symmetric.
    Kurtosis tells you the height and sharpness of the central peak, relative to that of a standard bell curve. (closer to 0 the better)

## the returns are not normally distributed

```python
#lets look at returns again 
plt.plot(returns)
plt.xlabel("Time")
plt.ylabel("returns")
plt.show()
#note that there is massive flux in magnitude and positivity of returns
#big volatility = volatility clustering (to try and group things that make sense together)
```

## Volatility clustering 

```python
#A closer look at the 20-days rolling standard deviation will make it look less like a heart attack graph
#20 day rolling = 20day groups 
plt.plot(returns.rolling(20).std()) 
plt.xlabel("Time")
plt.ylabel("20-days rolling standard deviation")
plt.show()
```
