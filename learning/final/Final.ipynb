```python
#save NA rows
dfNA = df[df.isnull().any(axis=1)]
#dfNA.count()
#drop NA for real DF
df = df.dropna()
df.count()
```

#### lets check that mobile and desktop likes add to total likes

```python
checkDF = df[df['likes'] != df['mobile_likes']+df['www_likes']]
display(checkDF)
```

it looks like the www_likes are simply filled to 0. lets fill them so it matches totalLikes

```python
df['www_likes'] = df.apply(lambda row: row['likes']-row['mobile_likes'], axis=1)
checkDF = df[df['likes'] != df['mobile_likes']+df['www_likes']]
checkDF.count()
```

lets do the same check and fill for likes received

```python
checkDF = df[df['likes_received'] != df['mobile_likes_received']+df['www_likes_received']]
checkDF.count() #=> 4
df['www_likes_received'] = df.apply(lambda row: row['likes_received']-row['mobile_likes_received'], axis=1)
checkDF = df[df['likes_received'] != df['mobile_likes_received']+df['www_likes_received']]
checkDF.count()
```

## Age group analysis

```python
print(df['age'].min(), df['age'].max())
```

lets use 10 as lower bound and 120 as upper bound to include the min maxes

```python
labels = [str(11+x)+"-"+str(20+x) for x in range(0, 110, 10)]
df['age_group'] = pd.cut(df['age'], bins = np.arange(10, 121, 10), labels = labels, right=True)

plt.subplots(figsize=(10,8))
sns.histplot(df, x = 'age_group', multiple="dodge", shrink = 0.8)
plt.show()
```

a majority of users are younger than 40 with the largest age group being 21-30 followed by 10-20 year olds.  
  
what age in 21-30 has to most users?

```python
df1=df[(df['age']>=21)&(df['age']<=30)]
df2=df1['age'].value_counts().reset_index().rename(columns={'age':'count','index':'age'})
df2.sort_values('age',ascending=True,inplace=True)
sns.barplot(x='age',y='count',data=df2,color='grey') #23
plt.show()
```

23 year olds are the most common age in the most common age group

```python
df1 = df['age'].value_counts().reset_index().rename(columns={'age':'count','index':'age'})
plt.subplots(figsize=(25,10))
sns.barplot(x='age',y='count',data=df1,color='grey') #23
plt.show()
```

## Activity analysis

```python
inactive_users = df.query('friend_count == 0 and friendships_initiated == 0 and likes == 0')
sns.histplot(data = inactive_users, x = 'gender', shrink = 0.8)
plt.show()
```

```python
df['activity_per_day'] = (df['friendships_initiated']/df['tenure'] + df['likes']/df['tenure']).replace([np.nan, np.inf, -np.inf], 0)
#activities_per_day is built from how many friendships are initiatied per day + likes per day
plt.subplots(figsize = (10,8))
sns.barplot(data = df,x = 'age_group',y = 'activity_per_day')
plt.show()
```

```python
f, ax = plt.subplots(12,figsize=(10, 35))
ageGroups = df.groupby('age_group')
axCt=0
for ageGp, ageGpDF in ageGroups:
    sns.barplot(data=ageGpDF, x='gender',y='activity_per_day', ax=ax[axCt]).set_title('age group: '+ageGp)
    axCt+=1
sns.barplot(data=df, x='gender', y='activity_per_day', ax=ax[axCt]).set_title('Total')
plt.show()
```

Females in age groups 10-80 are more active than males, while Males in age groups 80-120 are more active than females.  
In total, females are more active than males
